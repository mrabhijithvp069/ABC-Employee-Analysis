```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/ABC Company.xlsx - Sheet1.csv")
```

```python
# Correct the data in the "height" column by replacing it with random numbers between 150 and 180.
```

```python
df["height"] = np.random.randint(150, 181, size=len(df))
```

```python
#Determine the distribution of employees across each team and calculate the percentage split relative to the total number of employees.
```

```python
team_counts = df["Team"].value_counts()
team_percent = (team_counts / len(df)) * 100
print("Employees per team:\n", team_counts)
print("\nPercentage of employees per team:\n", team_percent)
sns.barplot(x=team_counts.index, y=team_counts.values)
plt.title("Employees per Team")
plt.xticks(rotation=90)
plt.show()
```

```python
#Some teams have more employees than others. The percentage split shows which teams dominate and which are smaller. This helps management see where most staff are concentrated.
```

```python
#Segregate employees based on their positions within the company.
```

```python
position_counts = df["Position"].value_counts()
sns.barplot(x=position_counts.index, y=position_counts.values)
plt.title("Employees per Position")
plt.xticks(rotation=30)
plt.show()
```

```python
#Employees are spread across different roles. Junior roles make up the bulk of the workforce, while senior roles are fewer.
```

```python
#Identify the predominant age group among employees.
```

```python
bins = [18, 25, 35, 45, 55, 65]
labels = ["1824", "2534", "3544", "4554", "5564"]
df["age_group"] = pd.cut(df["Age"], bins=bins, labels=labels)
age_group_counts = df["age_group"].value_counts()
sns.barplot(x=age_group_counts.index, y=age_group_counts.values)
plt.title("Employees per Age Group")
plt.show()
```

```python
#The most common age group is 2534 years. This suggests the company has a young workforce. Smaller older age groups may fewer senior employees
```

```python
#Discover which team and position have the highest salary expenditure.
```

```python
salary_team = df.groupby("Team")["Salary"].sum()
salary_position = df.groupby("Position")["Salary"].sum()

sns.barplot(x=salary_team.index, y=salary_team.values)
plt.title("Salary by Team")
plt.xticks(rotation=90)
plt.show()

sns.barplot(x=salary_position.index, y=salary_position.values)
plt.title("Salary by Position")
plt.xticks(rotation=90)
plt.show()
```

```python
#Certain teams contribute the most to salary costs. Senior positions like managers account have the highest salary.
```

```python
#Investigate if there's any correlation between age and salary, and represent it visually.
```

```python
print("Correlation:", df[["Age","Salary"]].corr().loc["Age","Salary"])
sns.scatterplot(x="Age", y="Salary", data=df)
plt.title("Age vs Salary")
plt.show()
```

```python
#The scatterplot shows whether salary increases with age. A positive correlation means older employees generally earn more. Weak correlation suggests salary depends more on role than age.
```
